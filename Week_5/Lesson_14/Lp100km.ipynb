```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
car_data = pd.read_csv('Lp100km.csv')
car_data.head()
```

```python
# Remove the rows with missing values in horsepower
car_data = car_data.loc[car_data["horsepower"] != "?"]
car_data.head()
```

```python
# Set the 'car name' as our index
car_data = car_data.set_index("car name")
# car_data.head()

# Remove the 'origin' column
del car_data["origin"]
car_data.head()
```

```python
# Convert the "horsepower" column to numeric so the data can be used
car_data["horsepower"] = car_data["horsepower"].astype(int)
car_data.dtypes
```

```python
# Create a scatter plot which compares L/100km to horsepower
filtered_df = car_data[["L/100km", "horsepower"]]
figure_1 = filtered_df.plot(kind="scatter", x="horsepower", y="L/100km", grid=True, figsize=(8,6), title="L/100km vs. Horsepower")
figure_1.set_xlabel("Horsepower")
figure_1.set_ylabel("Liters used (per 100km)")
fig = figure_1.figure
fig.savefig("L100km_vs._Horsepower.png")
plt.show()
```
