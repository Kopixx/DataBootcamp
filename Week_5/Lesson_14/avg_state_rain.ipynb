```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Using MatplotLib to Chart a DataFrame

```python
# Load in csv
rain_df = pd.read_csv("avg_rain_cities.csv")
rain_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(rain_df))
tick_locations = [i for i in x_axis]
```

```python
# Create a plot that includes a list indicating where to write x labels and set figure size to adjust for space
plt.bar(x_axis, rain_df["Millimetres"], color="red", alpha=0.75)
plt.xticks(tick_locations, rain_df["LOCATION"], rotation=90)

# Set x and y limits
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, max(rain_df["Millimetres"])+50)

# Set a Title and labels
plt.title("Average Amount of Rainfall in Australian Cities")
plt.ylabel("Average Rainfall (mm)")
plt.xlabel("Cities")

# Save our graph and show the grap
plt.tight_layout()
plt.savefig("avg_state_rain_output.png")
plt.show()
```

### Using Pandas to Chart a DataFrame

```python
# Filter the DataFrame down only to those columns to chart
rain_filt_df = rain_df[["Millimetres", "LOCATION"]]
# rain_filt_df.head()

# Set the index to be "State" so they will be used as labels
rain_filt_df = rain_filt_df.set_index("LOCATION")
rain_filt_df.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
rain_filt_df.plot(kind="bar", figsize=(8,6))

# Set a title for the chart
plt.title("Average Rainfall per Australian City")
```

```python
# Pandas can also plot multiple columns if the DataFrame includes them
multi_plot = rain_df.plot(kind="bar", figsize=(8,6))

# PandasPlot.set_xticklabels() can be used to set the tick labels as well
multi_plot.set_xticklabels(rain_df["LOCATION"], rotation=45)

#Set the title and axis labels
plt.title("Average Rainfall per Australian City")
plt.ylabel("Average Rainfall (mm)")
plt.xlabel("Cities")

#Save and display the plot
plt.tight_layout()
plt.savefig("avg_state_rain_pandas_output.png")
plt.show()
```
