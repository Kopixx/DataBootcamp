```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
# Import the WDI dataset, drop missing data
wdi_data = pd.read_csv('WDI_2018.csv')
wdi_data = wdi_data.dropna()
wdi_data.head()
```

```python
# Determine which pairs of factors are correlated. 
income = wdi_data.iloc[:,1]
alcohol = wdi_data.iloc[:,8]
correlation = st.pearsonr(income, alcohol)

#Construct Plot
plt.scatter(income, alcohol)
plt.title("Income vs. Alcohol")
plt.xlabel("Income per Capita")
plt.ylabel("Alcohol Consumption Per Person Per Year (L)")
plt.text(45000, 19, f'r = {round(correlation[0], 3)} (weak)')

#Save and Display Plot
plt.savefig("Correlation Scatterplot.png")
plt.show()
```
